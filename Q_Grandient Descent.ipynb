```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

先設 Loss Funtion 為 cos(x) + 9*x**3 + 4*x**2 - 8*x + 7 ，並畫出圖

```python
x = np.arange(-10,10,0.001)

def LF(x):
    return 9*x**3 + 4*x**2 - 8*x + 7


plt.plot(x,LF(x),'k')
plt.title('Loss Funtion')
plt.grid()
```

以不同的起始點開始慢慢迭代到 local minimum

```python
def dLF(x):
    return 27*x**2 + 8*x - 8

plt.plot(x,LF(x),'k')
plt.grid()
initial = 10 #起始點
# 畫出起始點
plt.scatter(initial,LF(initial))

iteration = 300 #迭代次數
lr = 0.0001 #學習速率
for i in range(iteration):
    initial = initial - lr*dLF(initial)
    plt.scatter(initial,LF(initial))
```

```python
plt.plot(x,LF(x),'k')
plt.grid()
initial = -2.5 #起始點
# 畫出起始點
plt.scatter(initial,LF(initial))

iteration = 200 #迭代次數
lr = 0.01 #學習速率
for i in range(iteration):
    initial = initial + lr*dLF(initial)
    plt.scatter(initial,LF(initial))
```
